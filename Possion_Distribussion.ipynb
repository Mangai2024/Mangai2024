**Poisson Distribution**

```python
The Poisson distribution is a probability distribution used to model the number of events that
 occur within a fixed interval of time or space, provided these events happen at a constant
average rate and independently of each other.
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Parameters for Poisson distribution
lambda_value = 5 # Average number of Dosa's per 10 minutes
size = 1000  # Number of random samples

# Generate random samples from a Poisson distribution
poisson_samples = np.random.poisson(lam=lambda_value, size=size)

# Plot histogram using seaborn
sns.histplot(poisson_samples, bins=10, kde=False, color='skyblue', stat='probability')
plt.xlabel('Number of Dosa(k)')
plt.ylabel('Probability')
plt.title(f'Histogram of Poisson Distribution (Î»={lambda_value})')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

Scenario:
A call center receives an average of 5 calls per hour. The manager wants to calculate:

The probability of receiving exactly 3 calls in an hour.
The probability of receiving fewer than 4 calls in an hour.
The probability of receiving 7 or more calls in an hour.

```python
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt
import seaborn as sns
# Parameters for the Poisson distribution
lmbda = 5  # Average number of calls per hour

# 1. Probability of exactly 3 calls
probability_exact_3 = poisson.pmf(3, lmbda)
print(f"Probability of exactly 3 calls in an hour: {probability_exact_3:.4f}")

# 2. Probability of fewer than 4 calls
probability_fewer_than_4 = poisson.cdf(3, lmbda)
print(f"Probability of fewer than 4 calls in an hour: {probability_fewer_than_4:.4f}")

# 3. Probability of 7 or more calls
probability_7_or_more = 1 - poisson.cdf(6, lmbda)
print(f"Probability of 7 or more calls in an hour: {probability_7_or_more:.4f}")
```
