```python
#STEP 7: Text Chunking (Preparation for GenAI / RAG)
texts = df['ingredients_text_clean'].tolist()
```

```python
def chunk_text(text, chunk_size=200):
    return [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]
```

```python
sample_chunks = chunk_text(texts[0])
sample_chunks
```

```python
#STep 8
!pip install sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
sample_texts = df['ingredients_text_clean'].head(10).tolist()
```

```python
embeddings = model.encode(sample_texts)

len(embeddings), len(embeddings[0])
```

```python
#STEP 9: Store Embeddings in FAISS & Perform Similarity Search
!pip install faiss-cpu
```

```python
import faiss
import numpy as np

embedding_dim = embeddings.shape[1]
index = faiss.IndexFlatL2(embedding_dim)

index.add(np.array(embeddings))
```

```python
query = "products with high sugar content"
query_embedding = model.encode([query])
```

```python
D, I = index.search(query_embedding, k=3)
I
```

```python
for idx in I[0]:
    print(sample_texts[idx])
```

```python
#STEP 10: Generate Answer using LLM (RAG Final Step)
from transformers import pipeline

qa_pipeline = pipeline(
    "text2text-generation",
    model="google/flan-t5-base"
)
```

```python
context = " ".join([sample_texts[idx] for idx in I[0]])
```

```python
prompt = f"""
Context:
{context}

Question:
{query}

Answer:
"""
```

```python
response = qa_pipeline(prompt, max_length=200)
response[0]['generated_text']
```
